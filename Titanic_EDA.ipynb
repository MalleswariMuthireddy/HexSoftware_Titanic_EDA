```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional nice plotting style
sns.set(style="whitegrid")

# Show all columns when printing DataFrames
pd.set_option("display.max_columns", None)
```

```python
#Load the Titanic CSV
df = pd.read_csv(r"C:\Users\india\OneDrive\Desktop\hex\train.csv")
print("Shape:", df.shape)
df.head()
```

```python
# Quick look at data
print("First 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())

print("\nRandom sample:")
display(df.sample(5, random_state=1))
```

```python
# Info and missing values
print("DataFrame info:")
df.info()

print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
#Summary statistics for numeric and object columns
print("Numeric summary:")
display(df.describe())

print("\nCategorical counts (example):")
display(df.select_dtypes(include=['object']).nunique())
```

```python
# Cleaning missing values
df_clean = df.copy()

# Fill Age with median
df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].median())

# Fill Embarked with mode
df_clean['Embarked'] = df_clean['Embarked'].fillna(df_clean['Embarked'].mode()[0])

# Drop Cabin column (many missing)
if 'Cabin' in df_clean.columns:
    df_clean = df_clean.drop(columns=['Cabin'])

# Verify
print("Missing values after cleaning:")
print(df_clean.isnull().sum())
```

```python
# Feature engineering
df_clean['FamilySize'] = df_clean['SibSp'] + df_clean['Parch'] + 1
df_clean['IsAlone'] = (df_clean['FamilySize'] == 1).astype(int)

# Extract title from Name (Mr, Mrs, Miss, etc.)
df_clean['Title'] = df_clean['Name'].str.extract(r',\s*([^\.]+)\.', expand=False).str.strip()
# group rare titles into 'Other'
title_counts = df_clean['Title'].value_counts()
rare_titles = title_counts[title_counts < 10].index
df_clean['Title'] = df_clean['Title'].replace(rare_titles, 'Other')

# Convert some columns to category dtype to save memory and for nicer analysis
for col in ['Pclass', 'Sex', 'Embarked', 'Title']:
    df_clean[col] = df_clean[col].astype('category')

df_clean.head()
```

```python
# Aggregations and groupby (future-proof, no warnings)
print("Overall survival rate:", df_clean['Survived'].mean())

print("\nSurvival rate by Sex:")
display(df_clean.groupby('Sex', observed=True)['Survived'].mean())

print("\nSurvival rate by Pclass:")
display(df_clean.groupby('Pclass', observed=True)['Survived'].mean())

print("\nAverage age by Pclass:")
display(df_clean.groupby('Pclass', observed=True)['Age'].mean())
```

```python
# Survival count
plt.figure(figsize=(5,4))
sns.countplot(x='Survived', data=df_clean)
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.show()
```

```python
# Survival by Sex
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df_clean)
plt.title('Survival by Sex')
plt.show()
```

```python
# Survival by Passenger Class
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=df_clean)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(8,4))
sns.histplot(df_clean['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
# Boxplot of Fare by Pclass
plt.figure(figsize=(8,4))
sns.boxplot(x='Pclass', y='Fare', data=df_clean)
plt.title('Fare distribution by Class')
plt.show()
```

```python
# Safe Correlation Heatmap

# Automatically select only numeric columns
numeric_df = df_clean.select_dtypes(include=['number'])  # includes int and float

# Check if there are numeric columns
if numeric_df.shape[1] == 0:
    print("No numeric columns found to plot correlation heatmap.")
else:
    plt.figure(figsize=(8,6))
    sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Correlation Heatmap")
    plt.show()
```

```python
# Survival by Title
plt.figure(figsize=(8,4))
sns.barplot(x='Title', y='Survived', data=df_clean, errorbar=None)
plt.title('Survival rate by Title')
plt.show()

# Survival by FamilySize
plt.figure(figsize=(8,4))
sns.barplot(x='FamilySize', y='Survived', data=df_clean, errorbar=None)
plt.title('Survival rate by Family Size')
plt.show()
```
